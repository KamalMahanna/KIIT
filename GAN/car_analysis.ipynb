```python
from sklearn.preprocessing import LabelEncoder
```

```python
leBrand = LabelEncoder()
```

```python
df['Brand'].unique()
```

```python
df['Brand'] = leBrand.fit_transform(df['Brand'])
```

```python
leModel = LabelEncoder()
df['Model'] = leModel.fit_transform(df['Model'])
leFuel_Type = LabelEncoder()
df['Fuel_Type'] = leFuel_Type.fit_transform(df['Fuel_Type'])
leTransmission = LabelEncoder()
df['Transmission'] = leTransmission.fit_transform(df['Transmission'])
leOwner_Type = LabelEncoder()
df['Owner_Type'] = leOwner_Type.fit_transform(df['Owner_Type'])
```

```python
df
```

```python
cont_cols = ['Year','Kilometers_Driven','Mileage','Engine','Power','Price']
```

```python
Q1 = df[cont_cols].quantile(0.25)
Q3 = df[cont_cols].quantile(0.75)
IQR = Q3-Q1
```

```python
df[((df[cont_cols]<(Q1-1.5*IQR)) | (df[cont_cols]>(Q3+1.5*IQR))).any(axis=1)]
```

```python
df_clean = df[~((df[cont_cols]<(Q1-1.5*IQR)) | (df[cont_cols]>(Q3+1.5*IQR))).any(axis=1)]
df_clean
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_clean.corr(),annot=True,fmt=".2f")
```
