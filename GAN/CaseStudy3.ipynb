```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataSets/loan_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
cont_cols = [col for col in df.columns if df[col].dtype != 'object' if col != 'loan_status']
cont_cols
```

```python
for i in cont_cols:
    plt.boxplot(x=df[i])
    plt.title(i)
    plt.show()
```

```python
# checking boxplot wrt loan_status
for i in cont_cols:
    sns.boxplot(data=df, y=i, hue='loan_status')
    plt.title(i)
    plt.show()
```
